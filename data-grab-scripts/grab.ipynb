```python
import requests
from bs4 import BeautifulSoup

card_list_url = "https://hololive-official-cardgame.com/cardlist/cardsearch"

# get max page from card list page
def get_max_page():
    response = requests.get(card_list_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find script tags containing max_page
    scripts = soup.find_all('script')
    max_page = 1
    
    for script in scripts:
        if script.string and 'max_page' in script.string:
            # Extract max_page using string operations
            script_content = script.string
            max_page_line = [line for line in script_content.split('\n') if 'max_page' in line]
            if max_page_line:
                # Extract the number from the line "var max_page = 57;"
                max_page = int(max_page_line[0].split('=')[1].strip('; '))
                break
    
    return max_page


def get_card_ids_list():
    card_list_by_page_url = "https://hololive-official-cardgame.com/cardlist/cardsearch_ex?page="
    max_page = get_max_page()
    card_ids_list = []
    for page in range(1, max_page + 1):
        response = requests.get(card_list_by_page_url + str(page))
        soup = BeautifulSoup(response.text, 'html.parser')
        for card in soup.find_all('a'):
            href = card.get('href')
            if href and '?id=' in href:
                # Extract the ID number from the href
                id_part = href.split('?id=')[1]
                # If there are additional parameters, take only the ID number
                if '&' in id_part:
                    card_id = id_part.split('&')[0]
                else:
                    card_id = id_part
                card_ids_list.append(card_id)
    return card_ids_list

print("Debug:", get_card_ids_list())
```

```python
card_info_url = "https://hololive-official-cardgame.com/cardlist/?id="

sample_card_ids = ['580', '565', '652', '578', '581', '612', '645', '662', '168']

def get_card_info(card_id):
    response = requests.get(card_info_url + card_id)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Initialize card info dictionary
    card_info = {'id': card_id}
    
    # Get card name
    try:
        card_name = soup.find('h1', class_='name').text.strip()
        card_info['name'] = card_name
    except:
        card_info['name'] = None
    
    # Get card text if available
    try:
        card_text = soup.find('div', class_='card_text').text.strip()
        card_info['text'] = card_text
    except:
        card_info['text'] = None
    
    # Find the txt-Inner div that contains all the card details
    txt_inner = soup.find('div', class_='txt-Inner')
    if not txt_inner:
        return card_info
    
    # Extract info from the main info section
    info_div = txt_inner.find('div', class_='info')
    if info_div:
        # Process main dl elements
        dl_elements = info_div.find_all('dl')
        for dl in dl_elements:
            dt_elements = dl.find_all('dt')
            dd_elements = dl.find_all('dd')
            
            for i in range(len(dt_elements)):
                if i < len(dd_elements):
                    key = dt_elements[i].text.strip()
                    
                    # Handle special case for the color attribute which has an image
                    if key == '色':
                        img = dd_elements[i].find('img')
                        if img and 'alt' in img.attrs:
                            card_info[key] = img['alt']
                    # Handle special case for Tags which have links
                    elif key == 'タグ':
                        tags = []
                        for tag_link in dd_elements[i].find_all('a'):
                            tags.append(tag_link.text.strip())
                        card_info[key] = tags
                    else:
                        # Regular text content
                        card_info[key] = dd_elements[i].text.strip()
    
    # Extract Arts information
    arts_div = txt_inner.find('div', class_='sp arts')
    if arts_div:
        arts_name = None
        arts_value = None
        
        # Try to extract the arts name and value
        p_tags = arts_div.find_all('p')
        if len(p_tags) > 1:
            spans = p_tags[1].find_all('span')
            if spans:
                arts_content = spans[0].text.strip()
                # Extract arts name and value (damage)
                parts = arts_content.split('　')
                if len(parts) > 1:
                    arts_name = parts[0]
                    arts_value = parts[1]
        
        card_info['アーツ'] = {
            'name': arts_name,
            'value': arts_value
        }
    
    # Extract Extra information
    extra_div = txt_inner.find('div', class_='extra')
    if extra_div:
        p_tags = extra_div.find_all('p')
        if len(p_tags) > 1:
            card_info['エクストラ'] = p_tags[1].text.strip()
    
    # Extract Oshi Skill information
    oshi_div = txt_inner.find('div', class_='oshi skill')
    if oshi_div:
        skill_p = oshi_div.find_all('p')
        if len(skill_p) > 1:
            skill_text = skill_p[1].text.strip()
            
            # Extract skill name from span if available
            skill_name = None
            skill_span = skill_p[1].find('span')
            if skill_span:
                skill_name = skill_span.text.strip()
                # Remove skill name from full text to get just the effect
                skill_text = skill_text.replace(skill_name, "", 1)
            
            card_info['推しスキル'] = {
                'name': skill_name,
                'effect': skill_text
            }
    
    # Extract SP Oshi Skill information
    sp_div = txt_inner.find('div', class_='sp skill')
    if sp_div:
        skill_p = sp_div.find_all('p')
        if len(skill_p) > 1:
            skill_text = skill_p[1].text.strip()
            
            # Extract skill name from span if available
            skill_name = None
            skill_span = skill_p[1].find('span')
            if skill_span:
                skill_name = skill_span.text.strip()
                # Remove skill name from full text to get just the effect
                skill_text = skill_text.replace(skill_name, "", 1)
            
            card_info['SP推しスキル'] = {
                'name': skill_name,
                'effect': skill_text
            }
    
    # Extract illustrator information
    illustrator_div = txt_inner.find('div', class_='illustrator')
    if illustrator_div:
        # Get illustrator name
        ill_name_p = illustrator_div.find('p', class_='ill-name')
        if ill_name_p:
            span = ill_name_p.find('span')
            if span:
                card_info['イラストレーター'] = span.text.strip()
        
        # Get card number
        number_p = illustrator_div.find('p', class_='number')
        if number_p:
            span = number_p.find('span')
            if span:
                card_info['カードナンバー'] = span.text.strip()
    
    # Process for サポート・イベント cards which have ability text
    ability_text = None
    dt_ability = info_div.find('dt', string='能力テキスト') if info_div else None
    if dt_ability:
        dd_ability = dt_ability.find_next_sibling('dd')
        if dd_ability:
            ability_text = dd_ability.text.strip()
            card_info['能力テキスト'] = ability_text
    
    return card_info

# Example function to get info for all cards and save to JSON
def get_multiple_cards_info(card_ids, is_sample=False):
    import json
    import time
    
    all_cards = []
    total_cards = len(card_ids)
    
    for i, card_id in enumerate(card_ids):
        print(f"Processing card {i+1}/{total_cards}: ID {card_id}")
        
        try:
            card_info = get_card_info(card_id)
            all_cards.append(card_info)
            
            # Save intermediate results every 50 cards
            if (i + 1) % 50 == 0:
                with open(f'cards_batch_{(i+1)//50}.json', 'w', encoding='utf-8') as f:
                    json.dump(all_cards[-50:], f, ensure_ascii=False, indent=2)
            
            # Be nice to the server
            time.sleep(1)  # Delay between requests to avoid overwhelming the server
            
        except Exception as e:
            print(f"Error processing card ID {card_id}: {str(e)}")
    
    # Save final results
    if is_sample:
        with open('sample_cards.json', 'w', encoding='utf-8') as f:
            json.dump(all_cards, f, ensure_ascii=False, indent=2)
    else:
        with open('all_cards.json', 'w', encoding='utf-8') as f:
            json.dump(all_cards, f, ensure_ascii=False, indent=2)
    
    return all_cards

# debug
# print("Debug:", get_card_info('565'))

# Sample Cards
get_multiple_cards_info(sample_card_ids, is_sample=True)
```
