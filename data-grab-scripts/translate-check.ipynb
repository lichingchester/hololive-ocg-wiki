```python
import json

ja_file='../locales/ja.json'
translated_file='../locales/tc.json'

# Load ja data
try:
    with open(ja_file, 'r', encoding='utf-8') as f:
        ja_data = json.load(f)
except FileNotFoundError:
    print(f"Error: Input file '{ja_file}' not found.")
except json.JSONDecodeError:
    print(f"Error: Input file '{ja_file}' contains invalid JSON.")

# Load translated data
try:
    with open(translated_file, 'r', encoding='utf-8') as f:
        translated_data = json.load(f)
except FileNotFoundError:
    print(f"Error: Input file '{translated_file}' not found.")
except json.JSONDecodeError:
    print(f"Error: Input file '{translated_file}' contains invalid JSON.")


# Compare card IDs between the two data sources
ja_card_ids = set()
translated_card_ids = set()

# Extract card IDs from the dictionary structure
if 'cards' in ja_data:
    ja_card_ids = set(ja_data['cards'].keys())
else:
    print("'cards' key not found in ja_data")

if 'cards' in translated_data:
    translated_card_ids = set(translated_data['cards'].keys())
else:
    print("'cards' key not found in translated_data")

print(f"ja_card_ids: {ja_data['cards'].keys()}")
print(f"translated_card_ids: {translated_data['cards'].keys()}")

for card_id in translated_card_ids:
    if card_id not in ja_card_ids:
        print(f"ID {card_id} is missing in ja data")
        continue  # Skip further comparisons if card_id is not in ja_data
    
    # Check if 'illustrator' key exists in both datasets before comparing
    ja_illustrator = ja_data['cards'][card_id].get('illustrator')
    translated_illustrator = translated_data['cards'][card_id].get('illustrator')

    if ja_illustrator is not None and translated_illustrator is not None:
        if ja_illustrator != translated_illustrator:
            print(f"ID {card_id} has different illustrator in ja data")
    elif ja_illustrator is None and translated_illustrator is not None:
        print(f"ID {card_id} has illustrator in translated data but not in ja data")
    elif ja_illustrator is not None and translated_illustrator is None:
        print(f"ID {card_id} has illustrator in ja data but not in translated data")

# Find missing IDs in each dataset
# missing_in_translated = ja_card_ids - translated_card_ids
# missing_in_ja = translated_card_ids - ja_card_ids

# # Print results
# if missing_in_translated:
#     print(f"IDs missing in translated data: {missing_in_translated}")
# else:
#     print("No IDs missing in translated data")

# if missing_in_ja:
#     print(f"IDs missing in ja data: {missing_in_ja}")
# else:
#     print("No IDs missing in ja data")
```
